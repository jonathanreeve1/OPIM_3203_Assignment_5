## Question 2. (20 points in total) Merging the two datasets or DataFrames and processing the merged DataFrame

Question 2 contains tasks 2a through 2e that you need to complete.

```python
# 2a. (6 points) Write Python code in this code cell
# to join the two DataFrames, apple_app_store_1_df and apple_app_store_2_df, through merging.
# In other words, we need to add the price values from apple_app_store_2_df
# to apple_app_store_1_df that contains more information of the apps.

# Please use some function in pandas and do a left join,
# using apple_app_store_1_df as the reference DataFrame.
# (as apple_app_store_1_df has more information compared to apple_app_store_2_df)
# And please assign the resulting merged DataFrame to a variable, called apple_app_store_merge_df.

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above
# about generating the DataFrame apple_app_store_1_df and the DataFrame apple_app_store_2_df
# before running this code cell.

apple_app_store_merge_df = pd.merge(apple_app_store_1_df, apple_app_store_2_df, how = 'left', on = 'app_name')
apple_app_store_merge_df
```

```python
# 2b. (2 points) Write Python code in this code cell
# to show all the unique price values in the resulting DataFrame apple_app_store_merge_df
# (hint: you can use the unique() method)

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# before running this code cell.

apple_app_store_merge_df['price_in_usd'].unique()
```

```python
# 2c. (6 points) Write Python code in this code cell
# to iterate over each value in the price_in_usd column of the DataFrame apple_app_store_merge_df,
# and determine whether an app is free or paid, based on its price.
# Store each free or paid value into a list called list_free_or_paid.

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# before running this code cell.

list_free_or_paid = []
for x in apple_app_store_merge_df['price_in_usd']:
  free_or_paid = 'Free'
  if x > 0:
    free_or_paid = 'Paid'
  list_free_or_paid.append(free_or_paid)
```

```python
# 2d. (4 points) Write Python code in this code cell
# to add a new column, called type, to the DataFrame apple_app_store_merge_df,
# and list_free_or_paid is assigned to the type column of the DataFrame apple_app_store_merge_df

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# before running this code cell.

apple_app_store_merge_df['type'] = list_free_or_paid
```

```python
# 2e. (2 point) Write Python code in this code cell
# to show all the unique category names in the DataFrame, apple_app_store_merge_df

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# before running this code cell.

apple_app_store_merge_df['category'].unique()
```

## Question 3. (62 points + 3 bonus points in total) Data visualization using the merged DataFrame

Question 3 contains tasks 3a through 3g that you need to complete.

```python
# 3a. (12 points) Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a pie plot for each type of apps (free vs. paid),
# so that the proportion of each type of apps can be visualized.

# You need to calculate the number of apps in each type (free vs. paid),
# use the color green for the slice of free apps, and use red for the slice of paid apps,
# and add a title to the plot, which is Proportion of free vs. paid apps in Apple app store

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

free_mask = apple_app_store_merge_df['type'] == "Free"
apple_app_store_merge_df_free = apple_app_store_merge_df[free_mask]
paid_mask = apple_app_store_merge_df['type'] == "Paid"
apple_app_store_merge_df_paid = apple_app_store_merge_df[paid_mask]

num_free = len(apple_app_store_merge_df_free.index)
num_paid = len(apple_app_store_merge_df_paid.index)

num_type_app = [num_free, num_paid]
type_app = ["Free", "Paid"]
color_slice = ["green","red"]

plt.figure()

plt.pie(num_type_app, labels = type_app, colors = color_slice, autopct = '%1.2f%%')
plt.title(" Proportion of free vs. paid apps in Apple app store")

plt.show()
```

```python
# 3b. (18 points) Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a bar plot
# that shows the total number of apps in each of the three categories:
# Education, Social Networking, Sports

# You need to calculate the total number of apps in each category,
# use the color blue for each bar, with a width of 0.3 in the plot,
# and add a title to the plot, which is Number of apps in three categories: Education, Social Networking, Sports

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

education_mask = apple_app_store_merge_df['category'] == 'Education'
apple_app_store_merge_df_education = apple_app_store_merge_df[education_mask]
num_education = len(apple_app_store_merge_df_education.index)

social_mask = apple_app_store_merge_df['category'] == 'Social Networking'
apple_app_store_merge_df_social = apple_app_store_merge_df[social_mask]
num_social = len(apple_app_store_merge_df_social.index)

sports_mask = apple_app_store_merge_df['category'] == 'Sports'
apple_app_store_merge_df_sports = apple_app_store_merge_df[sports_mask]
num_sports = len(apple_app_store_merge_df_sports.index)

num_category_app = [num_education, num_social, num_sports]
category_app = ['Education', 'Social Networking', 'Sports']

plt.figure()
plt.bar(category_app, num_category_app, color = 'blue', width = 0.3)
plt.xlabel('Category')
plt.ylabel('Number of apps')
plt.title('Number of apps in three categories: Education, Social Networking, Sports')
plt.show
```

```python
# 3c. (8 points) Then let's plot some distributions of ratings.
# Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a histogram
# for ratings of the apps in the Education category.

# You need to use the color blue for the bars in the plot,
# and add a title to the plot, which is Histogram of rating, education category

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.
education_mask = apple_app_store_merge_df["category"] == "Education"
apple_app_store_merge_df_education = apple_app_store_merge_df[education_mask]
rating_education = apple_app_store_merge_df_education["rating"]

plt.figure()
plt.hist(rating_education, bins = 10, color = "blue")
plt.title('Histogram of rating, Education category')
plt.show()
```

```python
# 3d. (8 points) Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a box plot
# for ratings of the apps in the Education category.

# You need to add a title to the plot, which is Box plot of rating, education category

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

education_mask = apple_app_store_merge_df["category"] == "EDUCATION"
apple_app_store_merge_df_education = apple_app_store_merge_df[education_mask]
rating_education = apple_app_store_merge_df_education["rating"]

plt.figure()
plt.boxplot(rating_education)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Box plot of rating, Education category')
plt.xticks([1], ["EDUCATION"])
plt.show()
```

```python
# 3e. (3 bonus points)
# What if we want a box plot that can put together the distributions of ratings
# for all categories in the data from apple_app_store_merge_df?

# Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a box plot
# that shows distributions of ratings for the categories in apple_app_store_merge_df.
# In total, apple_app_store_merge_df has 23 categories,
# and thus, the plot should have 23 boxes.
# Each box corresponds to a distribution of the ratings in one category.

# You need to add a title to the plot, which is Box plot of rating vs. category

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

dict_category_num_app = {}
for unique_category_name in apple_app_store_merge_df['category'].unique():
  category_mask = apple_app_store_merge_df['category'] == unique_category_name
  apple_app_store_merge_df_category = apple_app_store_merge_df[category_mask]
  rating_app = len(apple_app_store_merge_df_category.index)

  dict_category_num_app[unique_category_name] = rating_app

num_category_app = dict_category_num_app.values()
category_app = dict_category_num_app.keys()

plt.figure()
plt.box()
plt.title('Box plot of rating vs. category')
plt.show()
```

```python
# 3f. (8 points) Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a scatter plot
# that shows the relationship between app rating (put on the y axis of the plot)
# and number of reviews (put on the x axis of the plot)

# You need to use the color blue for the dots in the plot,
# and add a title to the plot, which is Scatter plot of rating vs. number of reviews

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

num_review_for_scatter_plot = apple_app_store_merge_df["num_review"]
rating_for_scatter_plot = apple_app_store_merge_df["rating"]

plt.figure(figsize=(12,8))
plt.scatter(num_review_for_scatter_plot, rating_for_scatter_plot, color = "blue")
plt.xlabel('Number of reviews')
plt.ylabel('Rating')
plt.title('Scatter plot of rating vs. number of reviews')
plt.show()
```

```python
# 3g. (8 points) Write Python code in this code cell
# to operate on the DataFrame apple_app_store_merge_df
# and use matplotlib.pyplot (as plt) to generate a scatter plot
# that shows the relationship between app rating (put on the y axis of the plot)
# and price (put on the x axis of the plot)

# You need to use the color green for the dots in the plot,
# and add a title to the plot, which is Scatter plot of rating vs. price

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame apple_app_store_merge_df
# and importing the libraries such as pandas and matplotlib.pyplot
# before running this code cell.

# We can first get data for price and ratings
price_for_scatter_plot = apple_app_store_merge_df["price_in_usd"]
rating_for_scatter_plot = apple_app_store_merge_df["rating"]

plt.figure(figsize=(12,8))
plt.scatter(price_for_scatter_plot, rating_for_scatter_plot, color = "green")
plt.xlabel('Price')
plt.ylabel('Rating')
plt.title('Scatter plot of rating vs. price')
plt.show()
```
